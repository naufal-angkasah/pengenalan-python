## Operator
Operator = Operator digunakan untuk melakukan operasi pada variabel dan nilai <br>
Macam-macam Operator sebagai berikut:
- Operator Aritmatika (Arithmetic Operators)
- Operator Perbandingan (Comparison (Relational) Operators)
- Operator Penugasan (Assignment Operators)
- Operator Logika (Logical Operators)
- Operator Bitwise (Bitwise Operators)
- Operator Keanggotaan (Membership Operators)
- Operator Identitas (Identity Operators)

### Operator Aritmatika (Arithmetic Operators)
`+, -, *, /, **, %, //`

```python
a = 10
b = 5

# Penjumlahan (+)
print("Penjumlahan:",a+b)

# Pengurangan (-)
print("Pengurangan:",a-b)

# Perkalian (*)
print("Perkalian:",a*b)

# Pembagian (/)
print("Pembagian:",a/b)

# Pangkat (**)
print("Pangkat:",a**b)

# Sisa bagi / Modulus (%)
print("Modulus",a%b)

# Pembagian Bulat (//)
print("Pembagian Bulat:", a//b)
```

### Operator Perbandingan (Comparison (Relational) Operators)
`==, !=, >, <, >=, <=, <>`

Operasi Perbandingan dapat bekerja syntax "literal"

```python
a = 10
b = 5

# Sama dengan (==)
print("Sama dengan:",a==b)

# Tidak sama dengan (!=)
print("Tidak sama dengan:",a!=b)

# Lebih dari (>)
print("Lebih dari:",a>b)

# Kurang dari (<)
print("Kurang dari:",a<b)

# Lebih dari sama dengan (>=)
print("Lebih dari sama dengan:",a>=b)

# Kurang dari sama dengan (<=)
print("Kurang dari sama dengan:",a<=b)
```

### Operator Penugasan (Assignment Operators)
`=, +=, -=, *=, /=, **=, %=, //=`

```python
# Sama dengan (=)
a=b=c=d=e=f=g=h = 10
print("Sama dengan:",a)
print("Sama dengan:",b)
#dst

# Tambah sama dengan (+=)
a+=3
print("Tambah sama dengan:",a)

# Kurang sama dengan (-=)
b-=3
print("Kurang sama dengan:",b)

# kali sama dengan (*=)
c*=3
print("Kali sama dengan:",c)

# bagi sama dengan (/=)
d/=3
print("bagi sama dengan:",d)

# Pangkat sama dengan (**=)
e**=3
print("Pangkat sama dengan:",e)

# Modulus sama dengan (%=)
f%=3
print("Modulus sama dengan:",f)

# Pembagian bulat sama dengan (//=)
g//=3
print("Pembagian bulat sama dengan:",g)
```

### Operator Logika (Logical Operators)
`and, or , not`

```python
x = 5

# and
print("Operator and:", x > 3 and x < 10)

# or
print("Operator or:", x < 5 or x < 4)

# not
print("Operator not:", not(x < 5 and x < 10))
```

### Operator Bitwise (Bitwise Operators)
`&, |, ^, ~, >>, <<`

```python
a = 12  # dalam biner: 1100
b = 6  # dalam biner: 0110

# AND (&) - menghasilkan 1 jika kedua bit bernilai 1
print("AND:",a&b)
# 1100
# 0110
# 0100 adalah 4

# OR (|) - menghasilkan 1 jika salah satu bit bernilai 1
print("OR:",a|b)
# 1100
# 0110
# 1110 adalah 14

# XOR (^) - menghasilkan 1 jika bit berbeda
print("XOR:",a^b)
# 1100
# 0110
# 1010 adalah 10

# NOT (~) - membalik semua bit
print("NOT:",~b)
# 0000000000000110
# 1111111111111001 adalah -7

# Left Shift (<<) - geser bit ke kiri
print("<<:",b<<2)
# 0000000000000110
# 0000000000011000 adalah 24

# Right Shift (>>) - geser bit ke kanan
print(">>:",b>>2)
# 0000000000000110
# 0000000000000001 adalah 1
```

### Operator Keanggotaan (Membership Operators)
`in, not in`

```python
x = 1,2,3,4,5

# in
print("in:", 1 in x)

# not in
print("not in:", 10 not in x)
```

### Operator Identitas (Identity Operators)
`is, is not`

is atau is not dapat dioperasikan untuk membandingkan memory/object

```python
x = 1
y = 1
z = 2

# is
print(x is y)

# is not
print(x is not z)
```

## Looping / Perulangan

Looping = Perulangan for digunakan untuk mengiterasi melalui sebuah urutan (yang dapat berupa daftar, tuple, kamus, himpunan, atau string). Ini berbeda dengan kata kunci for dalam bahasa pemrograman lain, dan lebih mirip dengan metode iterator yang ditemukan dalam bahasa pemrograman berorientasi objek lainnya.<br>

Di dalam bahasa pemrograman Python pengulangan dibagi menjadi 3 bagian, yaitu :
- For Loop

### For Loop
For = Perulangan for digunakan untuk mengiterasi melalui sebuah urutan (yang dapat berupa daftar, tuple, kamus, himpunan, atau string).
<br>
Dengan perulangan for, kita dapat menjalankan serangkaian pernyataan, sekali untuk setiap item dalam daftar, tuple, himpunan, dan sebagainya.

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print("saya suka makan buah:",x)
```

```python
angka = [1,2,3,4,5]
for x in angka:
  print(x)
```

```python
sekolah = "MAN 1 Banda Aceh"
for i in sekolah:
    print(i)
```

### While Loop
While = Dengan perulangan while, kita dapat menjalankan serangkaian pernyataan selama suatu kondisi tetap benar.

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
count = 0
while (count < 6):
  print ("The count is: ", count)
  count = count + 1

print ("Good bye!")
```

### Nested Loop
nested loop = Bahasa pemrograman Python memungkinkan penggunaan satu lingkaran di dalam loop lain. Pengulangan ini  ditulis dengan menggunakan lebih dari satu fungsi loop bisa if dan juga while

```python
print("Tabel Perkalian")
for i in range(1, 6):  # Loop luar: angka 1-5
    for j in range(1, 6):  # Loop dalam: angka 1-5
        hasil = i * j
        print(f"{i} x {j} = {hasil:2d}", end="  |  ")
    print()  # Pindah ke baris baru
```

## Object Oriented Programming (OOP)
OOP = OOP adalah singkatan dari Pemrograman Berorientasi Objek. Python adalah bahasa pemrograman berorientasi objek, yang memungkinkan Anda untuk mengorganisir kode Anda menggunakan kelas dan objek untuk organisasi yang lebih baik dan kemudahan penggunaan ulang

```python
class Manusia:
    def __init__(self):
        self.warna_rambut = "putih"
        self.tinggi = 170
        self.berat = 60
```

```python
Budi = Manusia()
```

```python
print(Budi.warna_rambut)
print(Budi.tinggi)
print(Budi.berat)
```

```python
class Manusia:
    def __init__(self, warna_rambut, tinggi, berat=60 ):
        self.warna_rambut = warna_rambut
        self.tinggi = tinggi
        self.berat = berat

    def cat_rambutnya(self, warna):
        self.warna_rambut = warna
        print(f"rambutnya jadi warna: {warna}")
```

```python
Budi = Manusia("hitam", 167)
```

```python
print(Budi.warna_rambut)
print(Budi.tinggi)
print(Budi.berat)
```

```python
Budi.cat_rambutnya("pink")
```
